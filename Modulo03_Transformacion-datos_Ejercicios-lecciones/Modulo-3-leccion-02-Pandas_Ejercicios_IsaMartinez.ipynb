### Creación e indexación de Series
1. Crea una serie con numeros correlativos entre 1 y 12 inclusive (puedes hacerlo como lista o diccionario): 

    - Cambia el índice de la serie anteriormente creada para que comience con el numero 20.


    - Nuevamente cambia el índice de la serie, esta vez con una lista que haga referencia a los meses del año (tenemos 12 elementos como la cantidad de meses del año).

    - Muestra los valores que están en la posición 3 y 8 de mi serie.

    - Muestra los valores que están etiquetados en enero y agosto.
  
    - Muestra sus valores, indices, forma, tamaño y el tipo de elementos de la serie.

```python
#Creo serie como lista
serie_lista=pd.Series(range(1,13))
print(f'La serie desde una lista es:\n{serie_lista}\n')
```

```python
#Cambio el índice para que comience en el 20
serie_lista.index=range(20, 32)
print(f'La serie con índices desde el 20 es:\n{serie_lista}')
```

```python
#Cambio el índice para que sean los meses del año
serie_lista.index=['enero', 'febrero', 'marzo', 'abril', 'mayo', 'junio', 'julio', 'agosto', 'septiembre', 'octubre', 'noviembre', 'diciembre']
print(f'La serie con índices como meses del año es:\n{serie_lista}')
```

```python
#Muestro los valores que están en las posiciones 3 y 8
print(f'Los valores en las posiciones 3 y 8 son:\n{serie_lista[[3,8]]}')
```

```python
#Muestro los valores que están etiquetados como 'enero' y 'agosto'
print(f'Los valores etiquetados en enero y agosto son:\n{serie_lista[["enero","agosto"]]}')
```

```python
#Muestro valores, indices, forma, tamaño y tipo de datos
print(f'Los valores de la serie son:\n{serie_lista.values}\n')
print(f'Los índices de la serie son:\n{serie_lista.index}\n')
print(f'La forma de la serie es: {serie_lista.shape}\n')
print(f'El tamaño de la serie es: {serie_lista.size}\n')
print(f'El tipo de datos de la serie es: {serie_lista.dtype}')
```
