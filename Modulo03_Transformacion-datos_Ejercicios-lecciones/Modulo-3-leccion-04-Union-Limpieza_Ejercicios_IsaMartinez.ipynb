1. Unid los dos DataFrames en uno solo, usando el método que considereis más correcto. Querremos quedarnos con todos los productos que estén en los dos conjuntos de datos. 

```python
#Traigo el primer csv
df_rating_performance=pd.read_csv('rating-and-performance.csv')
df_rating_performance.head(1)
```

```python
#Traigo el segundo csv
df_insight=pd.read_csv('insight.csv')
df_insight.head(1)
```

```python
#Uno ambos csv con merge inner para quedarme sólo con los productos que estén presentes en ambos csv
#Uno por la columna index porque ambos la tienen
df_merged=pd.merge(left=df_rating_performance, right=df_insight, left_on='index', right_on='index', how='inner')
df_merged.head(2)
```

2. Explorad el nombre de las columnas, ¿hay alguna que este repetida o que no nos aporte información de valor? ¿Hay alguna columna que solo tenga un valor? En caso de ser así, eliminadlas. **NOTA** Hay tres columnas que cuando exploramos sus valores únicos y sus frecuencias ayer, nos deberíamos haber dado cuenta que solo tiene un valor.  

```python
#Veo si hay alguna columna con etiqueta duplicada
#1. Creo una variable para guardar las posibles columnas duplicadas
duplicated_colums=df_merged.columns[df_merged.columns.duplicated()]

#2. Hago un if: si hay algún duplicado
if len(duplicated_colums)>0:
    print(f'La cantidad de columnas con etiquetas duplicadas es: {duplicated_colums}')
else:
    print('No hay columnas con etiquetas duplicadas')
```

```python
#Quiero ver sólo las columnas con un sólo valor único para pensar si las puedo eliminar
#1. Creo variable para guardar las columnas y sus valores únicos
valores=df_merged.nunique()

#2. Creo filtro con la variable
valores_unicos=valores[valores == 1]

#3. Print
print(valores_unicos)
```

```python
#Elimino:
#title_orig porque necesitamos tener sólo un título
#merchant.id porque ya tenemos merchant_id que tiene la misma información
#currency_buyer porque sólo hay EUR
#theme porque sólo hay 'summer'
df_merged.drop(['title_orig', 'merchant.id', 'currency_buyer', 'theme'], axis=1, inplace=True)
df_merged.head(1)
```

3. Algunas de las columnas tienen `.` en sus nombres y en otras tenemos `_`.  Unifica el nombre de las columnas para que todas tengan "_". Además, todas deben estar en minúsculas. 

```python
df_merged.columns=[i.replace('.', '_').lower() for i in df_merged.columns]
df_merged.head(1)
```

4. En la columna de `merchant_info_subtitle`, si nos fijamos hay algunas celdas donde tenemos entre paréntesis el número total de calificaciones para el comerciante (que es la misma información que la que tenemos en la columna "merchant_rating_count"). Eliminad de la columna "merchant_info_subtitle" la parte del *string* de los paréntesis. Guardar el resultado de vuestra limpieza en una columna que se llame igual, es decir, "merchant_info_subtitle". 

```python
#Le digo que reemplace todo lo que esté encerrado entre paréntesis por nada
df_merged['merchant_info_subtitle']=df_merged['merchant_info_subtitle'].str.replace('\(.*\)', '', regex=True)
df_merged.head(5)
```

5. Si nos fijamos, en la primera fila antes solo teníamos un paréntesis, pero los hemos eliminado en el paso anterior, y ahora tenemos un *string* vacío, lo cual no tiene mucho sentido. En este ejercicio debereis reemplazar esos *strings* vacios por nulos de Pandas (**NOTA** estos deben ser indicados de la siguiente forma `np.nan`, y deberéis importar numpy). 

```python
df_merged['merchant_info_subtitle']=df_merged['merchant_info_subtitle'].replace('', np.nan)
df_merged.head(5)
```
