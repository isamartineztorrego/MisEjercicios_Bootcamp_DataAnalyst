Las variables que tendremos en este dataframe son:
Incluye 26 variables para cada una de las canciones recogidas de spotify. Estas variables se describen
brevemente a continuaci√≥n:

- `Track`: nombre de la canci√≥n, tal y como aparece en la plataforma Spotify.
- `Artist`: nombre del artista.
- `Url_spotify`: Url de la canci√≥n.
- `Album`: el √°lbum en el que se encuentra la canci√≥n en Spotify.
- `Album_type`: indica si la canci√≥n se publica en Spotify como single o dentro de un √°lbum.
- `Uri`: un enlace de Spotify utilizado para encontrar la canci√≥n a trav√©s de la API.
- `Danceability`: describe lo adecuada que es una canci√≥n para bailar bas√°ndose en una
combinaci√≥n de elementos musicales como el tempo, la estabilidad del ritmo, la fuerza del
comp√°s y la regularidad general. Un valor de 0,0 es el menos bailable y 1,0 el m√°s bailable.
- `Energy`: es una medida de 0,0 a 1,0 y representa una medida perceptiva de intensidad y
actividad. Normalmente, las pistas energ√©ticas son r√°pidas, ruidosas y ruidosas. Por ejemplo, el
death metal tiene mucha energ√≠a, mientras que un preludio de Bach tiene una puntuaci√≥n baja
en la escala. Las caracter√≠sticas perceptivas que contribuyen a este atributo incluyen el rango
din√°mico, el volumen percibido, el timbre, la velocidad de inicio y la entrop√≠a general.
- `Key`: la clave de la pista. Los n√∫meros enteros se asignan a los tonos utilizando la notaci√≥n
est√°ndar Pitch Class. Por ejemplo, 0 = C, 1 = CÙÄÄÉ/DÙÄÄÑ, 2 = D, y as√≠ sucesivamente. Si no se detecta
ninguna tonalidad, el valor es -1.
- `Loudness`: la sonoridad global de una pista en decibelios (dB). Los valores de sonoridad se
promedian en toda la pista y son √∫tiles para comparar la sonoridad relativa de las pistas. La
sonoridad es la cualidad de un sonido que es el principal correlato psicol√≥gico de la fuerza f√≠sica
(amplitud). Los valores suelen oscilar entre -60 y 0 db.
- `Speechiness`: detecta la presencia de palabras habladas en una pista. Cuanto m√°s
exclusivamente hablada sea la grabaci√≥n (por ejemplo, un programa de entrevistas, un
audiolibro, poes√≠a), m√°s se acercar√° a 1,0 el valor del atributo. Los valores superiores a 0,66
describen pistas que probablemente est√©n compuestas en su totalidad por palabras habladas.
Los valores entre 0,33 y 0,66 describen pistas que pueden contener tanto m√∫sica como voz, ya
sea en secciones o en capas, incluyendo casos como la m√∫sica rap. Los valores por debajo de
0,33 representan probablemente m√∫sica y otras pistas no habladas.
- `Acousticness`: una medida de confianza de 0,0 a 1,0 para determinar si la pista es ac√∫stica. 1,0
representa una confianza alta en que la pista es ac√∫stica.
- `Instrumentalness`: predice si una pista no contiene voces. Los sonidos "ooh" y "aah" se
consideran instrumentales en este contexto. Las pistas de rap o spoken word son claramente
"vocales". Cuanto m√°s se acerque el valor de instrumental a 1,0, mayor ser√° la probabilidad de
que la canci√≥n no contenga voces. Los valores superiores a 0,5 representan pistas
instrumentales, pero la confianza es mayor a medida que el valor se acerca a 1,0.
- `Liveness`: detecta la presencia de p√∫blico en la grabaci√≥n. Los valores m√°s altos representan
una mayor probabilidad de que la pista se haya interpretado en directo. Un valor superior a 0,8
proporciona una gran probabilidad de que la pista sea en directo.
- `Valence`: una medida de 0,0 a 1,0 que describe la positividad musical que transmite una pista.
Las pistas con una valencia alta suenan m√°s positivas (por ejemplo, felices, alegres, euf√≥ricas),
mientras que las pistas con una valencia baja suenan m√°s negativas (por ejemplo, tristes,
deprimidas, enfadadas).
- `Tempo`: el tempo global estimado de una pista en pulsaciones por minuto (BPM). En terminolog√≠a
musical, el tempo es la velocidad o el ritmo de una pieza determinada y se deriva directamente
de la duraci√≥n media de los tiempos.
- `Duration_ms`: duraci√≥n de la pista en milisegundos.
- `Stream`: n√∫mero de streams de la canci√≥n en Spotify.
- `Url_youtube`: url del v√≠deo enlazado a la canci√≥n en Youtube, si lo tiene.
- `Title`: t√≠tulo del videoclip en Youtube.
- `Channel`: nombre del canal que ha publicado el v√≠deo.
- `Views`: n√∫mero de vistas.
- `Likes`: n√∫mero de me gusta.
- `Comments`: n√∫mero de comentarios.
- `Description`: descripci√≥n del v√≠deo en Youtube.
- `Licensed`: Indica si el v√≠deo representa contenido con licencia, lo que significa que el contenido
fue subido a un canal vinculado a un socio de contenido de Youtube y luego reclamado por dicho
socio.
- `official_video`: valor booleano que indica si el v√≠deo encontrado es el v√≠deo oficial de la
canci√≥n.

- Se quiere estudiar si existe una relaci√≥n entre la cantidad de visitas que reciben los
videos de m√∫sica de un artista en una plataforma de streaming y la cantidad de likes que
estos videos obtienen. Por lo tanto, queremos contestar a la siguiente pregunta: ¬øSon
aquellos artistas que tienen m√°s visitas los que m√°s likes tienen? Para solucionar este
ejercicio deber√©is:

    - Agrupa por artista y calcular la media de visitas y de likes.
    - Quedate con los 10 artistas que m√°s visitas han tenido.
    - Quedate con los 10 artistas que m√°s likes han tenido.
    - BONUS Haced lo mismo para los datos de tipo "single".

```python
#Creo una variable para cada groupby: uno con las medias de visitas y de likes
views_mean=df_spotify.groupby('artist')['views'].mean().sort_values(ascending=False).reset_index()
likes_mean=df_spotify.groupby('artist')['likes'].mean().sort_values(ascending=False).reset_index()
```

```python
#Los 10 artistas que m√°s visitas han tenido
top_10_views=views_mean.head(10)
top_10_views
```

```python
#Los 10 artistas que m√°s likes han tenido
top_10_likes=likes_mean.head(10)
top_10_likes
```

#### üîù A la pregunta *¬øSon aquellos artistas que tienen m√°s visitas los que m√°s likes tienen?*, si quiero comparar los resultados de los 2 top_10, utilizo `set` e `intersection`.

```python
#1. Saco los nombres de los artistas de cada top_10. Utilizo set para convertir las listas de nombres de artistas en conjuntos, lo cual facilita la comparaci√≥n
artist_top_10_views=set(top_10_views['artist'])
artist_top_10_likes=set(top_10_likes['artist'])
```

```python
#2. Encuentro artistas comunes con un .intersection(), que encuentra los elementos comunes entre los dos conjuntos
common_artist=artist_top_10_views.intersection(artist_top_10_likes)
print(f'Los artistas dentro de los dos top_10 son: {common_artist}')
```
